# Global COVID-19 Confirmed Cases Over Time
This notebook explores the global progression of the COVID-19 pandemic by visualizing the cumulative number of confirmed cases over time.The data is sourced from the 'covid-19 .csv'file, which povides daliy updates on COVID-19 statistics for countries and regions around the world.
By analyzing this data, we ami to understand how the virus spread globally and identify key inflection points in the pandemic timeline.

## Dataset Overview
The dataste contains the following key columns:
- Date: The date of the record
- Coubtry/Region: The country reporting the data
- State: Sub regional data
- Confirmed: Cumulative confirmedCOVID-19 cases
- Deaths:Cumulative deaths
- Recovered:Cumulative recoveries
  

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load dataset
df=pd.read_csv("covid19.csv")
df.info
df.describe()
df.columns
```

```python
# convert Date column to datetime format
df['Date']= pd.to_datetime(df['Date'])

# Group by data and sum confirmed cases globally
global_trend = df.groupby('Date')['Confirmed'].sum()
```

```python
# plot the global trend of confirmed cases
plt.figure(figsize=(18,8))
plt.plot(global_trend.index, global_trend.values,'y--', linewidth=2)
plt.title('Global COVID-19 Confirmed Cases Over Time')
plt.xlabel('Date')
plt.ylabel('confirmed cases')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show
```

## Observations 
 From the chart, we can observe:
 - A slow initial rise in cases, followedby exponential growth in early 2020
 - Periods of rapid acceleration, often correspondeing to global waves of infection.
 - A plateau of slowdown in growth may indicate the impact of public health interventions.
    This visualization provides a high level view of the pandemic's trajectory and can be extended to exlore country specific trends.
